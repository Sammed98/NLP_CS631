### Importing all the modules

```python
import re
import operator
from nltk.tokenize import sent_tokenize
from nltk.tokenize import word_tokenize
from nltk.util import ngrams
from sklearn.model_selection import train_test_split
import numpy as np
import math
import warnings
warnings.filterwarnings("always")
```

### Reading the The Adventures of Sherlock Holmes dataset

```python
with open('sherlock.txt', 'r') as f:
    AL = f.read()
#Reading the text file
```

### Sentence tokenizing of the data

```python
sent_list = sent_tokenize(AL)
#Sentence Tokenizing it. sent_list is a list of sentenses.
print("Number of sentences = " + str(len(sent_list)))
```

### Preprocessing of the data i.e Removeing the special characters, new line characters etc.

```python
#Removing the new line tags and special characters except spaces
for i in range(len(sent_list)):
    sent_list[i] = sent_list[i].replace("\n"," ").replace("\r"," ").replace("â€™","").replace("-"," ").replace("'","")
    sent_list[i] = re.sub('[^ A-Za-z0-9]',' ',sent_list[i])
    sent_list[i] = sent_list[i].lower()
```

### Splitting the dataset into training and testing data (80-20)

```python
sent_train, sent_test = train_test_split(sent_list, test_size=0.2, random_state=42)
print("Number of sentences in train dataset = " + str(len(sent_train)))
print("Number of sentences in test dataset = " + str(len(sent_test)))
```

### Word tokenizing, adding of opening and closing sentence tags  and converting all the words to lower case.

```python
#Word Tokenizing
word_list = []

for i in sent_list:
    word_list.append("<s>")
    word_list.extend(word_tokenize(i))
    word_list.append("</s>")
total_words = len(word_list)
print("Number of words (Tokens)= " + str(total_words))
```

### Calclating the number of unique words and their correponding frequencies

```python
#Calculating frequency of individual words after converting them to lower case
freq_words = {}
for i in word_list:
    i = i.lower()
    freq_words[i] = freq_words.get(i,0) + 1
print("Number of unique words =",len(freq_words))
freq_words
```

```python
def generateSent(model_name):
    if(model_name == "Unigram"):
        sentence = ""
        start = "<s> "
        sentence+=start
        wordlist = list(unigram_mle.keys())
        prob = list(unigram_mle.values())
        while(1):
            chosenarray = np.random.multinomial(1, prob)
            #print(np.where(chosenarray == 1))
            #print(int(np.where(chosenarray == 1)[0]))
            chosenword = wordlist[int(np.where(chosenarray == 1)[0])]
            if(chosenword == "<s>"):
                continue
            elif(chosenword == "</s>"):
                sentence = sentence + chosenword
                print(sentence, end ="")
                return
            else:
                sentence = sentence + chosenword + " " 
                
    elif(model_name == "Bigram"):
        sentence = " <s> "
        start = "<s>"
        while(1):
            nextWord = []
            prob = []
            for i in bigram_mle.keys():
                if(i[0] == start):
                    nextWord.append(i[1])
                    prob.append(bigram_mle[i])
            #print(sum(prob))
            chosenarray = np.random.multinomial(1, prob)
            chosenword = nextWord[int(np.where(chosenarray == 1)[0])]
            if(chosenword == "</s>"):
                sentence = sentence + chosenword
                print(sentence, end = "")
                return
            else:
                sentence = sentence + chosenword + " "
                start = chosenword
    elif(model_name == "Trigram"):
        #First generate the second word using the above logic
        sentence = "<s> "
        start_bi = "<s>"
        nextWord_bi = []
        prob_bi = []
        for i in bigram_mle.keys():
            if(i[0] == start_bi):
                nextWord_bi.append(i[1])
                prob_bi.append(bigram_mle[i])
        chosenarray_bi = np.random.multinomial(1, prob_bi)
        chosenword_bi = nextWord_bi[int(np.where(chosenarray_bi == 1)[0])]
        bigramTuple = (start_bi, chosenword_bi)
        #print(bigramTuple)
        sentence = sentence + chosenword_bi + " "
        while(1):
            nextWord_tri = []
            prob_tri = []
            for i in trigram_mle.keys():
                if(i[0] == bigramTuple[0] and i[1] == bigramTuple[1]):
                    nextWord_tri.append(i[2])
                    prob_tri.append(trigram_mle[i])
            chosenarray_tri = np.random.multinomial(1, prob_tri)
            chosenword_tri = nextWord_tri[int(np.where(chosenarray_tri == 1)[0])]
            bigramTuple = (bigramTuple[1],chosenword_tri )
            if(chosenword_tri == "</s>"):
                sentence = sentence + chosenword_tri
                print(sentence, end ="")
                return
            else:
                sentence = sentence + chosenword_tri + " "
    elif(model_name == "Quadgram"):
        #First generate the second word using the above logic
        sentence = "<s> "
        start_bi = "<s>"
        nextWord_bi = []
        prob_bi = []
        for i in bigram_mle.keys():
            if(i[0] == start_bi):
                nextWord_bi.append(i[1])
                prob_bi.append(bigram_mle[i])
        chosenarray_bi = np.random.multinomial(1, prob_bi)
        chosenword_bi = nextWord_bi[int(np.where(chosenarray_bi == 1)[0])]
        bigramTuple = (start_bi, chosenword_bi)
        sentence = sentence + chosenword_bi + " "
        nextWord_tri = []
        prob_tri = []
        for i in trigram_mle.keys():
            if(i[0] == bigramTuple[0] and i[1] == bigramTuple[1]):
                nextWord_tri.append(i[2])
                prob_tri.append(trigram_mle[i])
        chosenarray_tri = np.random.multinomial(1, prob_tri)
        chosenword_tri = nextWord_tri[int(np.where(chosenarray_tri == 1)[0])]
        trigramTuple = (start_bi, chosenword_bi,chosenword_tri)
        sentence = sentence + chosenword_tri + " "
        while(1):
            nextWord_quad = []
            prob_quad = []
            for i in quadgram_mle.keys():
                if(i[0] == trigramTuple[0] and i[1] == trigramTuple[1] and i[2] == trigramTuple[2]):
                    nextWord_quad.append(i[3])
                    prob_quad.append(quadgram_mle[i])
            #print(sum(prob_quad))
            chosenarray_quad = np.random.multinomial(1, prob_quad)
            chosenword_quad = nextWord_quad[int(np.where(chosenarray_quad == 1)[0])]
            trigramTuple = (trigramTuple[1],trigramTuple[2], chosenword_quad)
           # print(trigramTuple)
            if(chosenword_quad == "</s>"):
                sentence = sentence + chosenword_quad
                print(sentence, end ="")
                return
            else:
                sentence = sentence + chosenword_quad + " "
```

## Calculating the probability of a given sentence and a model
1. First preprocess the given sentence ie., removing the special characters, new line characters etc.
2. Word tokenize the sentence
3. Based on the given model calculate the tuples and append them into a list.
4. Calculate the probaility of each element of the tuple, apply log to base e and add them to give the probability of the sentence in log space.

### Note: Accepted model names are "Unigram", "Bigram", "Trigram"and "Quadgram".

```python
def logProbUnigram(words_list):
    log_sum_uni = 0 
    for i in words_list:
        #print(i, end = " ")
        try:
            log_sum_uni += math.log(unigram_mle[i])
        except:
            log_sum_uni+=0
    print("\n")
    print("Log space probability Unigram=",math.exp(log_sum_uni))
    
def logProbBigram(words_list):
    log_sum_bi = 0 
    for i in words_list:
        #print(i, end = " ")
        try:
            log_sum_bi += math.log(bigram_mle[i])
        except:
            log_sum_bi+=0
    print("\n")
    print("Log space probability Bigram =",math.exp(log_sum_bi))
    
def logProbTrigram(words_list):
    log_sum_tri = 0 
    for i in words_list:
        #print(i, end = " ")
        try:
            log_sum_tri += math.log(trigram_mle[i])
        except:
            log_sum_tri+=0
    print("\n")
    print("Log space probability Trigram =",math.exp(log_sum_tri))
def logProbQuadgram(words_list):
    log_sum_quad = 0 
    for i in words_list:
        #print(i, end = " ")
        try:
            log_sum_quad += math.log(quadgram_mle[i])
        except:
            log_sum_quad+=0
    print("\n")
    print("Log space probability Quadgram =",math.exp(log_sum_quad))
    
    
def Probability(sentence,model_name):
    #Prepocessing of sentence
    sentence = sentence.replace("\n"," ").replace("\r"," ").replace("â€™","").replace("-"," ")
    sentence = re.sub('[^ A-Za-z0-9]',' ',sentence)
    
    #Creating a list of words with the starting and ending tag.
    words_list =[] 
    words_list.append("<s>")
    words_list.extend(word_tokenize(sentence))
    words_list.append("</s>")
    for i in range(len(words_list)):
        words_list[i] = words_list[i].lower()
    if(model_name == "Unigram"):
        logProbUnigram(words_list)
    elif(model_name == "Bigram"):
        logProbBigram(ngrams(words_list,2))
    elif(model_name == "Trigram"):
        logProbTrigram(ngrams(words_list,3)) 
    elif(model_name == "Quadgram"):
        logProbQuadgram(ngrams(words_list,4))
    else:
        print("Unexpected input.")
    
sentence = "A slow and heavy step, which had been heard upon the stairs and in the passage, paused immediately outside the door. Then there was a loud and authoritative tap."
Probability(sentence, "Unigram")
Probability(sentence, "Bigram")
Probability(sentence, "Trigram")
Probability(sentence, "Quadgram")
```
